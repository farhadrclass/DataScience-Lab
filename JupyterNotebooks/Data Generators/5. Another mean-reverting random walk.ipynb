```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import math
import random
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
```

```python
Asset=1
Mean=0.15
Var=0.2
V=0.15
TimeStep=0.01
n=400 #number of sample generated 
```

```python
TimeArr=[]
TimeArr.append(0)
AssetArr=[]
AssetArr.append(Asset)
```

```python
for i in range(1,n):
    TimeArr.append(TimeArr[i-1]+TimeStep)
    dummy= AssetArr[i-1]+((V-Mean*AssetArr[i-1])*TimeStep)+(Var*(math.sqrt(TimeStep*AssetArr[i-1])))*(random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)+random.uniform(0, 1)-6)
    AssetArr.append(dummy)
# print(TimeArr)
# print(AssetArr)
```

```python
x = TimeArr
y = AssetArr

plt.scatter(x, y,alpha=0.5)
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Asset")
plt.legend(loc=2)
plt.show()
```
